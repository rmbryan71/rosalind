Plan:
1. Read in strings
2. Populate matrix with counts
3. Build consensus string with max values from matrix

```python
with open('rosalind_cons.txt') as f:
    read_data = f.read()
f.closed

def parse_fasta(s):
    results = {}
    strings = s.strip().split('>')
    
    for s in strings:
        if len(s) == 0:
            continue
            
        parts = s.split()
        label = parts[0]
        values = ''.join(parts[1:])
        
        results[label] = values
    return results

def build_matrix(input):
    labels, width = parse_fasta(input).items()
    height = 4
    profile = [[0 for x in range(width)] for y in range(height)]
    profile[0:0] = {'A'}
    for x,y in parse_fasta(input).items():
        i = 1 # counter for which string we're on.
        print (x,y)
        for z in y:
            print (z)
            if z == 'A':
                print ('recorded A')
                #profile[1,i] += 1
            elif z == 'C':
                print ('recorded C')
                #profile[2,i] += 1
            elif z == 'G':
                print ('recorded G')
                #profile[3:i] += 1
            elif z == 'T':
                print ('recorded T')
                #profile[4:i] += 1
        i += 1
    return profile
            
print (build_matrix(read_data))
```
