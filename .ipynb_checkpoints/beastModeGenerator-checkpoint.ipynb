```python
import random
import string
import itertools
```

```python
def makeDNA(length):
    result = ''.join(random.choice('ACGT') for _ in range(length))
    return result

def makeSeqName():
    result = ''.join(random.choice(string.ascii_uppercase) for _ in range(9))
    return str('Seq_' + result)

def makeRecord(length):
    return str('>' + makeSeqName() + '\n' + makeDNA(length) + '\n')
```

```python
def splc_beast():
    open('./datasets/rosalind_splc_beast.txt', 'w').close()
    with open('./datasets/rosalind_splc_beast.txt', 'w') as f:
        f.write(makeRecord(1000000))
        for x in range(1,10000):
            f.write(makeRecord(10))
    f.close
```

```python
splc_beast()
```
