Given: A DNA string s (of length at most 1 kbp) and a collection of substrings of s acting as introns. All strings are given in FASTA format.

Return: A protein string resulting from transcribing and translating the exons of s. (Note: Only one solution will exist for the dataset provided.)

```python
from Bio.Seq import Seq
from Bio import SeqIO

def read_fasta(fasta_file):
    """Read fasta file and return the list of records, consisting of
    description and sequence"""
    handle = open(fasta_file)
    records = list(SeqIO.parse(handle, "fasta"))
    handle.close()
    return [(rec.description, str(rec.seq)) for rec in records]

from Bio.Alphabet import generic_dna
from Bio.Alphabet import generic_rna

GENCODE = {
    'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
    'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
    'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
    'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',
    'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
    'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
    'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
    'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
    'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
    'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
    'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
    'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
    'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
    'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
    'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_',
    'TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W'}


def reverse_complement(seq):
    """Given a DNA sequence, return its reverse complement"""
    return str(Seq(seq).reverse_complement())

def transcribe(dna_seq):
    """Transcribe DNA sequence into mRNA sequence"""
    return str(Seq(dna_seq, generic_dna).transcribe())
    
def translate(rna_seq):
    """Translate RNA sequence to amino acid sequence"""
    mrna = Seq(rna_seq, generic_rna)
    return str(mrna.translate())
```

```python
def remove_substring(a,b):
    '''returns a with all instances of b removed'''
    return a.replace(b,'')

def splc(mode):
    if mode == 'sample':
        dataset = read_fasta('./datasets/rosalind_splc_example.txt')
    elif mode == 'rosalind':
        dataset = read_fasta('./datasets/rosalind_splc.txt')
    elif mode == 'beast':
        dataset = read_fasta('./datasets/rosalind_splc_beast.txt')
    else:
        print('Incorrect mode value. Valid entries are: sample, rosalind, beast.')
        
    str1 = str((dataset[0][1]))
    #print(str1)
    for x in range(1,len(dataset)):
        #print(translate(transcribe(str1.replace(str(dataset[x][1]),''))))
        str1 = str1.replace(str(dataset[x][1]),'')
    print(translate(transcribe(str1)))
    
%timeit splc('rosalind')
```
