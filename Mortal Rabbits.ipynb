```python
n = input("Enter months to simulate.")
m = input("Enter month of age when rabbits die.")

gens = [1, 1]  # First generation is initial adult, reproducing pair.


def mortal_rabbits(i, j):
    count = 2 
    while count < i:
        if count < j:
            gens.append(gens[-2] + gens[-1]) 
        elif (count == j or count == j+1):
           gens.append((gens[-2] + gens[-1]) - 1)
        else:
            gens.append((gens[-2] + gens[-1]) - (gens[-(j+1)]))            
        count += 1
    return (gens[-1])


print(mortal_rabbits(n, m))
```
