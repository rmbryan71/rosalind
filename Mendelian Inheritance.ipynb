```python
def prob(k, m, n):
    pop = k + m + n #Population
    prob = 0 #Probability of dominant allele
    
    # grid of possible results with likelihood of dominant allele offspring
    kk = 4/4
    km = 4/4
    kn = 4/4
    mk = 4/4
    mm = 3/4
    mn = 2/4
    nk = 4/4
    nm = 2/4
    nn = 0
    
    # k branch
    prob += k/pop * (k-1)/(pop-1) * kk
    prob += k/pop * m/(pop-1) * km
    prob += k/pop * n/(pop-1) * kn
    # m branch
    prob += m/pop * k/(pop-1) * mk
    prob += m/pop * (m-1)/(pop-1) * mm
    prob += m/pop * n/(pop-1) * mn
    # n branch
    prob += n/pop * k/(pop-1) * nk
    prob += n/pop * m/(pop-1) * nm

    return prob


print (prob(25, 22, 25))
```
