Given: A DNA string of length at most 1 kbp in FASTA format.

Return: The position and length of every reverse palindrome in the string having length between 4 and 12. You may return these pairs in any order.

```python
from Bio.Seq import Seq
from Bio import SeqIO

def read_fasta(fasta_file):
    '''Read fasta file and return the list of records, consisting of
    description and sequence'''
    handle = open(fasta_file)
    records = list(SeqIO.parse(handle, "fasta"))
    handle.close()
    return [(str(rec.seq)) for rec in records]
```

```python
def is_palindrome(input):
    seq = ''.join(input)
    length = len(seq)
    
    for i in range(0,length): #could stop half way
        #print(seq[i], seq[length-1-i])
        if not (seq[i] == seq[length-1-i]):
            return False
    return True
```

```python
def is_palindrome_compliment(input):
    seq = ''.join(input)
    length = len(seq)
    
    for i in range(0,length): #could stop half way
        #print(seq[i], seq[length-1-i])
        if not (seq[i] == compliment(seq[length-1-i])):
            return False
    return True
    
def compliment(input):
    if input == 'A' : return 'T'
    elif input == 'T': return 'A'
    elif input == 'G' : return 'C'
    elif input == 'C' : return 'G'
    else: return input
```

```python
# dataset = read_fasta("datasets/rosalind_revp_example.txt")
dataset = read_fasta("datasets/rosalind_revp.txt")
str1 = ''.join(str(dataset))
length = len(str1)
halfway = length//2

for i in range(0,length):
    for j in range(4,13):
        k = str1[i:i+j]
        if (i+j < length) and is_palindrome_compliment(k):
            print(i-1, j)
```
