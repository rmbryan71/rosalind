```python
def pop(n,k): # Population given 1 pair to start and n generations of k pairs of offspring
    if n < 2:
        return 1
    else:
        return (pop(n-1,k)+k * pop(n-2, k))

print(pop(4,3)) #base zero months, so enter 4 here to calculate 5 months worth of rabbits
    

        
```

```python
def pop2(n,k):
    a, b = 1, 1
    for i in range(2, n):
        a, b = b, k*a + b
    return b

print(pop2(5,3))
```

```python
def pop3(n,m):
    if n < 1:
        return 0
    if n < 2:
        return n
    else:
        return pop3(n-1,m) + (pop3(n-2,m)*2) - pop3(n-(m-1),m)
    
print(pop3(6,3))
```

```python
n = input("Enter months to simulate.")
m = input("Enter month of age when rabbits die.")
n = int(n)
m = int(m)

gens = [1, 1]  # First generation is initial adult, reproducing pair.


def mortal_rabbits(i, j):
    count = 2 
    while count < i:
        if count < j:
            gens.append(gens[-2] + gens[-1]) 
        elif (count == j or count == j+1):
           gens.append((gens[-2] + gens[-1]) - 1)
        else:
            gens.append((gens[-2] + gens[-1]) - (gens[-(j+1)]))            
        count += 1
    return (gens[-1])


print(mortal_rabbits(n, m))
```
